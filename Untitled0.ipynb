```python
# Step 2: Extract the ZIP
import zipfile
import os

zip_path = 'hierarchical modeling exercise.zip'  # This must match the uploaded filename
extract_path = 'hierarchical_exercise_extracted'

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

# See what files were extracted
os.listdir(extract_path)
```

```python
# Step 3: Load the dataset
import pandas as pd

csv_path = os.path.join(extract_path, 'student_scores.csv')
df = pd.read_csv(csv_path)

# Optional: View first few rows
df.head()
```

```python
# Step 4: Hierarchical modeling
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Aggregate to school-level summaries
school_summary = df.groupby('school_id').agg({
    'test_score': 'mean',
    'study_hours': 'mean',
    'school_funding': 'first',
    'urban': 'first'
}).reset_index()

# Standardize features
features = ['test_score', 'study_hours', 'school_funding', 'urban']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(school_summary[features])

# Perform hierarchical clustering
Z = linkage(X_scaled, method='ward')

# Plot dendrogram
plt.figure(figsize=(10, 6))
dendrogram(Z, labels=school_summary['school_id'].values)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('School ID')
plt.ylabel('Distance')
plt.show()
```
